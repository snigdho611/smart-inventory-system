```python
#function that takes a CSV file with transactions and cleans it

def clean_data(file):
    groceries_DataFrame = pd.read_csv(file)
    groceries_DataFrame = groceries_DataFrame.fillna(0)
    return groceries_DataFrame
```

```python
def unique_data(cleaned_data):
    groceries_unique = list()
    length = len(cleaned_data.index)

    for i in cleaned_data.index:
        for j in range(1, len(cleaned_data.iloc[i])):
            if cleaned_data.iloc[i][j] != 0 and cleaned_data.iloc[i][j] != '0':
                if cleaned_data.iloc[i][j] not in groceries_unique:
                    groceries_unique.append(cleaned_data.iloc[i][j])
                    
    groceries_unique_list = list()
    groceries_unique_list.append(['ID', 'Item', 'Price'])
    counter = 0
    for i in groceries_unique:
        groceries_unique_list.append([counter, i, str(round(random.random()*20, 2))])
        counter = counter + 1

    print(groceries_unique_list)
    return groceries_unique_list
```
